Create simple 2D images that demonstrate density distribution of core vs. cusp

```python
#load necessary packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Define density profiles
def rho_nfw(r,rs=2,p0=1):
    rho = p0 / ((r/rs)*(1+r/rs)**2)
    return rho

def rho_pis(r,rc=2,p0=1):
    rho = p0 * (1+(r/rc)**2)**(-1)
    return rho
```

```python
#get arrays for plotting
rad = np.linspace(0.01,2,100)
#p_nfw = rho_nfw(r)
#p_iso = rho_pis(r)

p0 = np.full(len(rad),1.)
rs = np.full(len(rad),3.)
rc = np.full(len(rad),1.)
p_nfw = (p0 / ((rad/rs)* (1+rad/rs)**2)) / 298.
p_iso = p0 * (1+(rad/rc)**2)**(-1)

print(min(p_nfw),min(p_iso))
print(max(p_nfw),max(p_iso))
```

```python
#make 2d plots
fig = plt.figure(figsize = (12,12))

ax = plt.subplot(1, 1, 1, projection='polar')

azm = np.linspace(0, 2 * np.pi)
r, th = np.meshgrid(rad, azm)
z_nfw = np.tile(np.log10(p_nfw), (r.shape[0], 1))

ax.pcolormesh(th, r, z_nfw)
ax.set_yticklabels([])
ax.set_xticklabels([])

plt.savefig('logdens_nfw.png')
```

```python
#make 2d plots
fig = plt.figure(figsize = (12,12))

ax = plt.subplot(1, 1, 1, projection='polar')

azm = np.linspace(0, 2 * np.pi)
r, th = np.meshgrid(rad, azm)
z_iso = np.tile(np.log10(p_iso), (r.shape[0], 1))

ax.pcolormesh(th, r, z_iso)
ax.set_yticklabels([])
ax.set_xticklabels([])
plt.savefig('logdens_iso.png')
```
